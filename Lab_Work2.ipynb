```python
win = 5
weights = np.arange(0.1, 0.6, 0.1)
def plot_graph(df):
    SMA5 = pd.DataFrame()
    SMA5['Adj Close'] = df['Adj Close'].rolling(window = 5).mean()
    SMA10 = pd.DataFrame()
    SMA10['Adj Close'] = df['Adj Close'].rolling(window = 10).mean()
    wma_company = df['Adj Close'].rolling(win).apply(
        lambda x: np.sum(weights*x)/weights.sum())
    
    plt.figure(figsize = (13,5))
    plt.plot(df['Adj Close'], label = 'Adj Close Price')
    plt.plot(SMA5['Adj Close'], label = 'SMA5')
    plt.plot(SMA10['Adj Close'], label = 'SMA10')
    plt.plot(list(range(len(wma_company))), wma_company,
             'r--', label=f'WMA: window size = 5')
    
    plt.xlabel('13 Dec 2021 - 12 Dec 2022')
    plt.ylabel('Adj Close Price USD ($)')
    plt.legend(loc = 'upper right')
    plt.show()
```

```python
plot_graph(df_TSLA)
```

```python
files = os.listdir('Stocks')
for file in files:
    df = pd.read_csv(f"Stocks/{file}", index_col = "Date")
    plot_graph(df)
    print(f'plot for the company: {file}')
```

### Применена стандартная скалярная нормализация для нормализации набора признаков.

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)
```

### Применение метода главных компонент (PCA)

```python
from sklearn.decomposition import PCA

pca = PCA()
X = pca.fit_transform(X)
```

### Дисперсия, вызванная каждой из главных компонент

```python
explained_variance = pca.explained_variance_ratio_
explained_variance
```
